```julia
function selectAction(board,d,γ,n)
    if d == 0 || numMoves(board) == 0
        return(0,scoreBoard(board))
    end
    astar, vstar = 0, -1000000
    for a in randperm(4)-1
        pboard = executeMove(a,board)
        if pboard != board
            v = 0
            nStates = countEmpty(pboard)
            for i in 1:n
                cboard = insertRandTile(pboard)
                ap, vp = selectAction(cboard,d-1,γ,n)
                v += (scoreBoard(cboard) + γ*vp)/n
                #v += (vp)/n
            end
            if v > vstar
                astar, vstar = a,v
            end
        end
    end
    return(astar, vstar)
end
```

```julia
function sparseSample(board::UInt64)
    return selectAction(board,4,.75,4)[1]
end
```

```julia
tic()
n = 100
s = zeros(n,2)
for i in 1:n
    tic()
    val, score = runGame(sparseSample, false)
    println(i, " ", val, " ", score)
    s[i,1] = val
    s[i,2] = score
    toc()
end
toc()
println(mean(s[:,2]))
println(std(s[:,2]))

s = convert(DataFrame, s)
CSV.write("sparseSample4-4.csv", s)
```
